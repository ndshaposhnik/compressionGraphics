```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 7))
prefix = 'nn_exp_losses/'
dimension = 15142970
experiments = ['top_05.csv', 'without_compression.csv']

with open('nn_exp_losses/without_compression.txt', "r") as f:
    data = list(map(float, f.readlines()))
    plt.semilogy(
        [dimension * k for k in range(1, len(data) + 1)],
        data,
        label='Without',
        linewidth=2
    )

with open('nn_exp_losses/top05/loss_history.txt', "r") as f:
    data = list(map(float, f.readlines()))
    plt.semilogy(
        [dimension * k // 2 for k in range(1, len(data) + 1)],
        data,
        label='Top 0.5',
        linewidth=2
    )

with open('nn_exp_losses/rand05_163.txt', "r") as f:
    data = list(map(float, f.readlines()))
    plt.semilogy(
        [dimension * k // 2 for k in range(1, len(data) + 1)],
        data,
        label='Rand 0.5',
        linewidth=2
    )

with open('nn_exp_losses/reduce_probability/loss_history.txt', "r") as f:
    data = list(map(float, f.readlines()))
    plt.semilogy(
        [dimension * k // 2 for k in range(1, len(data) + 1)],
        data,
        label='Reduce Probability 0.5',
        linewidth=2
    )

with open('nn_exp_losses/penalty.txt', "r") as f:
    data = list(map(float, f.readlines()))
    plt.semilogy(
        [dimension * k // 2 for k in range(1, len(data) + 1)],
        data,
        label='Penalty 0.5',
        linewidth=2
    )
    
with open('nn_exp_losses/topUnknown/loss_history.txt', "r") as loss, open('nn_exp_losses/topUnknown/transmitted_coordinates_history.txt', "r") as transmitted_coordinates:
    loss = list(map(float, loss.readlines()))
    transmitted_coordinates = list(map(float, transmitted_coordinates.readlines()))
    plt.semilogy(
        np.cumsum(transmitted_coordinates),
        loss,
        label='Top Unknown',
        linewidth=2
    )

with open('nn_exp_losses/marina/loss_history.txt', "r") as loss, open('nn_exp_losses/marina/transmitted_coordinates_history.txt', "r") as transmitted_coordinates:
    loss = list(map(float, loss.readlines()))
    transmitted_coordinates = list(map(float, transmitted_coordinates.readlines()))
    plt.semilogy(
        np.cumsum(transmitted_coordinates),
        loss,
        label='Marina',
        linewidth=2
    )

plt.legend(loc="best", fontsize=16)
plt.xlabel(r"Transmitted coordinates", fontsize=20)
plt.ylabel("loss", fontsize=20)
plt.xticks(fontsize=18)
_ = plt.yticks(fontsize=18)
```
