```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 7))
prefix = 'nn_exp_losses/'
dimension = 15142970
experiments = ['top_05.csv', 'without_compression.csv']

with open('nn_exp_losses/without_compression.txt', "r") as f:
    data = list(map(float, f.readlines()))
    plt.semilogy(
        [dimension * k for k in range(len(data))],
        data,
        label='Without',
        linewidth=2
    )

with open('nn_exp_losses/top05/loss_history.txt', "r") as f:
    data = list(map(float, f.readlines()))
    plt.semilogy(
        [dimension * k // 2 for k in range(len(data))],
        data,
        label='Top 0.5',
        linewidth=2
    )

with open('nn_exp_losses/rand05/loss_history.txt', "r") as f:
    data = list(map(float, f.readlines()))
    plt.semilogy(
        [dimension * k // 2 for k in range(len(data))],
        data,
        label='Rand 0.5',
        linewidth=2
    )


plt.legend(loc="best", fontsize=16)
plt.xlabel(r"Transmitted coordinates", fontsize=20)
plt.ylabel("loss", fontsize=20)
plt.xticks(fontsize=18)
_ = plt.yticks(fontsize=18)
```
